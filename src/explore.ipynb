```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv

# load the .env file variables
load_dotenv()

```

```python
client_id = os.environ.get('CLIENT_ID')
client_secret = os.environ.get('CLIENT_SECRET')
#print(client_id) solo para saber que funciona, no debemos publicar contraseñas
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

con = spotipy.Spotify(auth_manager = SpotifyClientCredentials(client_id=client_id,client_secret=client_secret))

id_show = '6U02wirujU8zNsDQcbPJhS'
id_artista = '790FomKkXshlbRYZFtlgla'

#response = con.show_episodes((id_show))
response = con.artist_top_tracks(id_artista)
print(response['tracks']) 

#utilizar jsonformatter para poner el texto en formato json

df_tracks = pd.DataFrame.from_records(response['tracks'])
print (df_tracks)
```

```python
diagrama_de_correlacion = sns.scatterplot(data=df_tracks,x='popularity',y='duration_ms')
fig = diagrama_de_correlacion.get_figure()
fig.savefig('scatter.png')
```
